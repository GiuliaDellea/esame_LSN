## Equilibrazione

La prima cosa da fare è implementare la lettura della posizione precendente e il riscalamento delle velocità per imporre la temperatura desiderata. Ho leggermente modificato il calcolo del riscalamento delle velocità rispetto a come veniva proposto nella consegna: consigliavano di calcolare $r(t + dt)$, da questo la velocità $v(t + dt/2)$ e poi trovare una nuova posizione iniziale compatibile con la temperatura desiderata. Ho notato che in realtà non è necessario calcolare $r(t + dt)$, infatti conosco già $r(t - dt)$ e $r(t + dt)$ e di conseguenza posso calcolare $v(t - dt/2)$ e da questa una nuova $r(t)$ compatibile con la temperatura target.

### Solido

Per l'equilibrazione del solido ho preferito partire da una temperatura superiore rispetto a quella desiderata. Partendo dalla configurazione fcc il sistema uscirà dalla condizione di equilibrio, diminuendo l'energia cinetica e di conseguenza la temperatura. In questo modo posso decidere di fare iniziare la prima simulazione a una temperatura più alta, scelta in modo da ottenere alla fine della simulazione la temperatura desiderata. Questo procedimento accorcia l'equilibrazione del solido, pertanto soltanto dal secondo tentativo di equilibrazione utilizzo il riscalamento delle velocità con una simulazione che comincia alla temperatura desiderata.

Nella seguente simulazione ho fatto partire la prima equilibrazione a una temperatura di $T = 1.5$ al posto di $T = 0.8$. Tutte le seguenti equilibrazioni sono state fatte con $1000$ passi dell'algoritmo di Verlet. La configurazione di partenza della prima è il cristallo fcc, mentre per le altre la configurazione iniziale è la configurazione finale della simulazione precedente.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
equi1 = np.loadtxt("../Esercizio4/output_temp.dat", delimiter=' ', unpack='true')
```

```python
equi2 = np.loadtxt("../Esercizio4/output_temp.dat", delimiter=' ', unpack='true')
```

```python
equi3 = np.loadtxt("../Esercizio4/output_temp.dat", delimiter=' ', unpack='true')
```

```python
# Prepara tre immagini
fig, ax = plt.subplots(1, 3, figsize=(15,5), tight_layout = True)

x = (np.arange(100) + 1) * 10
equi = [equi1, equi2, equi3]

# Fai i grafici
for i in range(3):
    ax[i].plot(x, equi[i])
    ax[i].set_xlabel("Passi Verlet")
    ax[i].set_ylabel("T")
    ax[i].set_title("Equilibrazione temperatura nel solido")
    ax[i].grid(True)
```

In questo modo si vede che soltanto con tre simulazioni ho raggiunto una buona equilibrazione.

## Calcolo dell'energia e della temperatura

Dopo aver descritto come fare la parte di equilibrazione, la simulazione è semplice. Ho aggiunto delle funzioni per fare le medie a blocchi, ovvero le solite per calcolare il valore di un blocco, fare le somme progressive e per calcolare gli errori. Nella simulazione utilizzo le unità di Lennard-Jones, di conseguenza volendo i risultati nel sistema internazionale devo fare le seguenti conversioni

$$V = \epsilon V^{*}, \quad K = \epsilon K^{*}, \quad E = \epsilon E^{*}, \quad T = \frac{\epsilon}{k_B} T^{*}$$

dove $\epsilon/k_B = 120 $ K e di conseguenza $\epsilon = 0.0103$ eV.

```python
# Prepara i fattori di conversione per le unità di misura
unitàE = 0.0103 # In eV
unitàT = 120 # In K
```

### Solido

Dopo aver equilibrato il sistema ho utilizzato per la simulazione $20$ blocchi ognuno con $1000$ step di Verlet.

```python
# Carica i dati
pot, pot_err = np.loadtxt("../Esercizio4/ave_epot.dat", delimiter=' ', usecols = (0,1), unpack='true') * unitàE
kin, kin_err = np.loadtxt("../Esercizio4/ave_ekin.dat", delimiter=' ', usecols = (0,1), unpack='true') * unitàE
ene, ene_err = np.loadtxt("../Esercizio4/ave_etot.dat", delimiter=' ', usecols = (0,1), unpack='true') * unitàE
tem, tem_err = np.loadtxt("../Esercizio4/ave_temp.dat", delimiter=' ', usecols = (0,1), unpack='true') * unitàT

# Prepara quattro immagini per gli istogrammi
fig, ax = plt.subplots(2, 2, figsize = (14, 8), tight_layout = True)
ax = ax.reshape(-1)

# Vettori
assex = [1000 * (i + 1) for i in range(20)]
valori = [pot, kin, ene, tem]
errori = [pot_err, kin_err, ene_err, tem_err]
simboli = ["U/N [eV]", "K/N [eV]", "E/N [eV]", "T [K]"]

# Plotta
for i in range(4):
    ax[i].errorbar(assex, valori[i], yerr = errori[i])
    ax[i].set_xlabel("Passi algoritmo Verlet")
    ax[i].set_ylabel(simboli[i])
    ax[i].grid(True)
    
ax[0].set_title("Energia potenziale per particella del solido")
ax[1].set_title("Energia cinetica per particella del solido")
ax[2].set_title("Energia totale per particella del solido")
ax[3].set_title("Temperatura del solido")
plt.show()
```

Dal grafico della temperatura possiamo vedere che il sistema ha mantenuto una temperatura leggermente superiore rispetto alla desiderata $T = 96$ K, ma il risultato è accettabile considerando la difficoltà di una equilibrazione molto precisa. Si può notare che mentre le incertezze sull'energia potenziale e sull'energia cinetica sono dell'ordine di $10^{-5}$, l'incertezza sull'energia totale è dell'ordine di $10^{-6}$. Questo conferma che l'algoritmo conserva in modo preciso l'energia totale.

### Gas

Per la simulazione del gas raggiungo l'equilibrazione e poi utilizzo le stesse condizioni delle precendenti.
